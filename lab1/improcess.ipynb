```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# open the image passed as an argument
file_name = 'parrots.jpg'
im = Image.open(file_name)

# convert to greyscale
im = im.convert('L')

# resize
im = im.resize((100, 100))

# load it's data into a numpy array and close the file
im_data = np.asarray(im, dtype=np.uint8)
im.close()
flat_im_data = im_data.flatten()

# ==== EXERCISE a ====
# sort the data
sorted_im_data = np.sort(flat_im_data)

# plot the data
plt.plot(sorted_im_data)
plt.ylabel('px intensity')
plt.show()
```

```python
# ==== EXERCISE b ====
plt.xlabel('px intensity')
plt.ylabel('px count')
plt.grid(True)
plt.hist(sorted_im_data, bins=32)
plt.show()
```

```python
# ==== EXERCISE c ====
im_height = im_data.shape[0]
im_width = im_data.shape[1]

lrq_lslice = slice(int(im_height / 2), im_height)
lrq_cslice = slice(int(im_width / 2), im_width)

im_crop = im_data[lrq_lslice, lrq_cslice]
plt.imshow(im_crop, cmap='gray')
plt.show()
```

```python
# ==== EXERCISE d ====
intensity_th = np.median(flat_im_data)
print(intensity_th)
```

```python
# ==== EXERCISE e ====
binary_im = np.array([1 if x > intensity_th else 0 for x in flat_im_data]).reshape(im_data.shape)
plt.imshow(binary_im, cmap='gray')
plt.show()
```

```python
# ==== EXERCISE f ====
intensity_av = np.floor(np.average(flat_im_data))
darkened_im = np.array([px - intensity_av if px > intensity_av else 0 for px in flat_im_data])
darkened_im = darkened_im.reshape(im_data.shape)
plt.imshow(darkened_im, cmap='gray')
plt.show()
```

```python
# ==== EXERCISE g ====
min_intensity = np.min(flat_im_data)
min_intens_pos = np.argwhere(im_data == min_intensity)
print(min_intens_pos)
```
