```python
import scipy
import scipy.ndimage as ndimage
import numpy
import matplotlib.pyplot as pyplot

text = ndimage.imread('./textura.png', mode='L')

filter = numpy.full((7, 7), 255)
filter[1:-1,1:-1] = 0
filter = text[0:7, 0:5]
filter = filter / numpy.sum(filter)
filter = filter - numpy.mean(filter)

filtered = ndimage.correlate(text, filter)

pyplot.figure()
pyplot.imshow(filtered, 'gray')
pyplot.figure()
pyplot.imshow(filter, 'gray')
pyplot.show()
```
